## Imputação de Valores

Deveria ser feita depois da separação entre treino e teste para evitar vazamento de dados...

```python
print(covid_df.isnull().sum())
```

```python
doencas = [
    "Pneumopatia", 
    "Nefropatia", 
    "Doença Hematológica", 
    "Distúrbios Metabólicos", 
    "Imunopressão", 
    "Obesidade", 
    "Outros", 
    "Cardiovasculopatia"
]
```

```python
ax = covid_df.groupby(["Nefropatia"]).size().plot(kind="bar")
for p in ax.patches:
    ax.annotate(
        np.round(p.get_height(),decimals=2), 
        (p.get_x()+p.get_width()/2., p.get_height()), 
        ha='center', 
        va='center', 
        xytext=(0, 5), 
        textcoords='offset points'
    )
```

Escolha de imputar os valores ausentes como `Não` 

```python
for doenca in doencas:
    covid_df[doenca].fillna("Não", inplace = True)
```

```python
ax = covid_df.groupby(["Pneumopatia"]).size().plot(kind="bar")
for p in ax.patches:
    ax.annotate(
        np.round(p.get_height(),decimals=2), 
        (p.get_x()+p.get_width()/2., p.get_height()), 
        ha='center', 
        va='center', 
        xytext=(0, 5), 
        textcoords='offset points'
    )
```

```python
ax = covid_df.groupby(["RA"]).size().plot(kind="bar")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), rotation=65)
```
