```python
import pandas as pd
```

```python
covid_df = pd.read_csv("../data/dados_df_imputados.csv", sep=",")
covid_df
```

```python
import numpy as np
```

```python
covid_df["tempoDeVida"] = pd.to_datetime(covid_df["Data do Óbito"], format="%Y/%m/%d", errors="coerce") - pd.to_datetime(covid_df["dataPrimeirosintomas"], format="%Y/%m/%d", errors="coerce") 
```

```python
covid_df[covid_df["tempoDeVida"].notnull()]
```

```python
np.sum(covid_df["tempoDeVida"].notnull())
```

```python
covid_df["tempoDeVidaInt"] = pd.to_numeric(covid_df['tempoDeVida'].dt.days, downcast='integer')
```

```python
covid_df.groupby(["tempoDeVidaInt"]).size()
```

```python
covid_df[(covid_df["tempoDeVidaInt"] > 0) & (covid_df["tempoDeVidaInt"] < 30)]
```

```python
covid_df["dataPrimeirosintomas"] = pd.to_datetime(covid_df["dataPrimeirosintomas"], format="%Y/%m/%d", errors="coerce")
```

```python
covid_df["dataPrimeirosintomas"].dt.year == 2020
```

```python
covid_df[(covid_df["tempoDeVidaInt"] > 0) & (covid_df["tempoDeVidaInt"] < 30) & (covid_df["dataPrimeirosintomas"].dt.year == 2020)].groupby(["tempoDeVidaInt"]).size().plot(kind="bar")
```

```python
covid_df[(covid_df["tempoDeVidaInt"] > 0) & (covid_df["tempoDeVidaInt"] < 30) & (covid_df["dataPrimeirosintomas"].dt.year == 2021)].groupby(["tempoDeVidaInt"]).size().plot(kind="bar")
```

```python
covid_df[(covid_df["tempoDeVidaInt"] > 0) & (covid_df["tempoDeVidaInt"] < 30) & (covid_df["dataPrimeirosintomas"].dt.year == 2022)].groupby(["tempoDeVidaInt"]).size().plot(kind="bar")
```
